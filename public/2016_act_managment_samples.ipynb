```python
import sys
import xlsxwriter
import pandas as pd
import sys
import psycopg2
import os
from dotenv import load_dotenv
from datetime import date

today = date.today()
load_dotenv()

#configurar la conexion a la base de datos
DB_USERNAME = os.getenv('DB_USERNAME')
DB_DATABASE = os.getenv('DB_DATABASE')
DB_PASSWORD = os.getenv('DB_PASSWORD')
DB_PORT = os.getenv('DB_PORT')
DB_HOST=os.getenv('DB_HOST')

# Conectar a DB
# Conectar a PostgreSQL
try:
    cnx = psycopg2.connect(
        user=DB_USERNAME,
        password=DB_PASSWORD,
        host=DB_HOST,
        port=DB_PORT,
        database=DB_DATABASE
    )
    print("Conexión exitosa")
except psycopg2.Error as e:
    print("Ocurrió un error al conectar a la base de datos:", e)

egresados=pd.read_sql("""select * from egresados  where act_suvery=1 """,cnx)
```

```python
egresados=pd.read_sql("""select * from egresados  where act_suvery=1 """,cnx)
print(len(egresados))
egresados=pd.read_sql("""select egresados.*,count(telefonos.id) as telefonos from egresados  left join telefonos on telefonos.cuenta=egresados.cuenta where act_suvery=1 group by egresados.id""",cnx)
print(len(egresados))
```

```python
carreras=egresados['carrera'].unique()
planteles=egresados['plantel'].unique()
carrera_meta=pd.read_sql('select * from carreras',cnx)
```

```python
list_carreras=carrera_meta.loc[(carrera_meta['clave_carrera'].isin(egresados['carrera'].unique()))&(carrera_meta['clave_plantel'].isin(egresados['plantel'].unique()))]
```

```python
writer = pd.ExcelWriter('storage/muestras16.xlsx', engine='xlsxwriter')

workbook = writer.book
a_color='#173d83'
#estilos----------------
negro_b = workbook.add_format({
    'bold': 2,
    'border': 0,
    'align': 'center',
    'valign': 'vcenter',
    'font_color': 'black',
    'font_size':13}) 
header_format = workbook.add_format({
    'bold': True,
    'bg_color': a_color,
    'text_wrap': True,
    'valign': 'top',
    'align': 'center',
    'border_color':'white',
    'font_color': 'white',
    'border': 1,
    'font_size':12})
blue_content = workbook.add_format({
    'border': 1,
    'align': 'center',
    'valign': 'vcenter',
    'font_color': 'black',
    'font_size':10,
    'border_color':a_color})
date_content = workbook.add_format({
    'border': 1,
    'align': 'center',
    'valign': 'vcenter',
    'font_color': 'black',
    'font_size':10,
    'border_color':a_color,
    'num_format': 'dd/mm/yy'})
date_content_bold = workbook.add_format({
    'align': 'center',
    'valign': 'vcenter',
    'font_color': 'black',
    'font_size':12,
    'bold': True,
    'num_format': 'dd/mm/yy'})
worksheet = workbook.add_worksheet()
worksheet.merge_range('C2:H3', 'PROGRAMA DE VINCULACION A EGRESADOS UNAM', negro_b)
worksheet.merge_range('C4:H4', 'ESTADO DE LAS MUESTRAS 2016', negro_b)
worksheet.insert_image("A1", "img/logoPVE.png",{"x_scale": 0.2, "y_scale": 0.2})
worksheet.merge_range('G6:H6',today, date_content_bold)

worksheet.write('B8','No.carrera',header_format)
worksheet.write('C8','No.plantel',header_format)
worksheet.write('D8','carrera',header_format)
worksheet.write('E8','plantel',header_format)
worksheet.write('F8','Poblacion',header_format)
worksheet.write('G8','Porcentaje',header_format)
worksheet.write('H8','Egresados sin telefono',header_format)
worksheet.write('I8','Porcentaje de esta muestra sin telefono',header_format)
worksheet.write('J8','Poblacion localizable via telefonica',header_format)
# worksheet.write('K8','Correo 2',header_format)
# worksheet.write('L8','Correo 3',header_format)
# worksheet.write('M8','Correo 4',header_format)
for i in range(len(list_carreras)):
    car=list_carreras['clave_carrera'].values[i]
    plant=list_carreras['clave_plantel'].values[i]
    eg=egresados.loc[(egresados['carrera']==car)&(egresados['plantel']==plant)]
    if(len(eg)>0):
        worksheet.write('B'+str(i+9),car,blue_content)
        worksheet.write('C'+str(i+9),plant,blue_content)
        worksheet.write('D'+str(i+9),list_carreras['carrera'].values[i],blue_content)
        worksheet.write('E'+str(i+9),list_carreras['plantel'].values[i],blue_content)
        worksheet.write('F'+str(i+9),len(eg),blue_content)
        worksheet.write('G'+str(i+9),str(round((len(eg)*100)/len(egresados),2))+"%",blue_content)
        worksheet.write('H'+str(i+9),len(eg.loc[eg['telefonos']==0]),blue_content)
        worksheet.write('I'+str(i+9),str(round((len(eg.loc[eg['telefonos']==0])*100)/len(eg),2))+"%",blue_content)
        worksheet.write('J'+str(i+9),len(eg.loc[eg['telefonos']>0]),blue_content)

worksheet.fit_to_pages(1, 1)  
workbook.close()
```
