```python
#historial de commits
#!/usr/bin/env python3
from git import Repo
from datetime import datetime, timedelta
from collections import defaultdict

def get_commits_by_author(repo_path, days=7):
    repo = Repo(repo_path)
    since_date = datetime.now() - timedelta(days=days)
    
    author_stats = defaultdict(int)
    for commit in repo.iter_commits(since=since_date):
        author_stats[commit.author.name] += 1
    
    return author_stats

if __name__ == "__main__":
    stats = get_commits_by_author(".", days=7)
    for author, count in sorted(stats.items(), key=lambda x: x[1], reverse=True):
        print(f"{author}: {count} commits")
```

```python
import subprocess
from datetime import datetime, timedelta
from collections import defaultdict

def report_commits_by_author(days=90, repo_path='.'):
    since_date = (datetime.now() - timedelta(days=days)).strftime("%Y-%m-%d")
    cmd = [
        "git", "-C", repo_path, "log",
        f"--since={since_date}",
        "--format=%an"      # Solo el nombre del autor
    ]
    result = subprocess.run(cmd, capture_output=True, text=True)
    if result.returncode != 0:
        print("Error ejecutando git log:", result.stderr)
        return {}
    

    
    authors = result.stdout.strip().split("\n")
    stats = defaultdict(int)
    for author in authors:
        if author:
            stats[author] += 1
    
    return stats

# Ejecutar
stats = report_commits_by_author(days=90)
for author, count in sorted(stats.items(), key=lambda x: x[1], reverse=True):
    print(f"{author}: {count} commits")
```
